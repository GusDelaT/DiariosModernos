```python
# Importar bibliotecas necesarias
import os
import pandas as pd
from PIL import Image
from pathlib import Path
import numpy as np
```

```python
# Ruta a la carpeta de imágenes en Drive
carpeta_imagenes = '/content/drive/MyDrive/PAIT/IMAGENES_130125_190125'

# Crear una lista de archivos de imagen automáticamente
lista_imagenes = list(Path(carpeta_imagenes).rglob("*.[jp][pn]g"))

# Crear una lista para almacenar la información de la imagen
datos_imagenes = []

# Iterar a través de las imágenes en la lista
for ruta_imagen in lista_imagenes:
    # Abrir la imagen usando Pillow
    with Image.open(ruta_imagen) as img:
        # Convertir la imagen a un array de NumPy
        imagen_array = np.array(img)

        # Obtener el nombre del archivo
        nombre_archivo = ruta_imagen.name

        # Agregar la información a la lista
        datos_imagenes.append([nombre_archivo, imagen_array])
```

```python
# Crear un DataFrame a partir de la lista
df_imagenes = pd.DataFrame(datos_imagenes, columns=['Nombre de archivo', 'Imagen'])

# Mostrar el DataFrame
df_imagenes
```
